```python
from timeit import Timer
import matplotlib.pyplot as plt
```

```python
def  test_concate(x):
    l = []
    for i in range(x):
        l = l + [i]
```

```python
def test_append(x):
    l = []
    for i in range(x):
        l.append(i)
```

```python
def test_comprehension(x):
    l = [ i for i in range(x) ]
```

```python
def test_list(x):
    l = list(range(x))
```

```python
if __name__ == "__main__" :
    t_concat = []
    t_append = []
    t_comprehension = []
    t_list = []
    x = range(1000,10000,1000)
    for i in x:
        
        
        t1 = Timer("test_concate(i)", "from __main__ import test_concate, i"  )
        #print ("Concat", t1.timeit(number=1000),"milliseconds")

        t2 = Timer("test_append(i)", "from __main__ import test_append; from __main__ import i")
        #print ("Append", t2.timeit(number=1000),"milliseconds")

        t3 = Timer("test_comprehension(i)", "from __main__ import test_comprehension, i")
        #print ("Comprehension", t3.timeit(number=1000),"milliseconds")

        t4 = Timer("test_list(i)", "from __main__ import test_list, i")
        #print ("List", t4.timeit(number=1000),"milliseconds")
    
        t_concat.append(t1.timeit(number=100))
        t_append.append(t2.timeit(number=100))
        t_comprehension.append(t3.timeit(number=100))
        t_list.append(t4.timeit(number=1000))

    
```

```python
    plt.figure()
    plt.subplot(411)
    plt.plot(x,t_concat, label="List Concatination Operation")
    plt.legend()
    plt.subplot(412)
    plt.plot(x,t_append, label="List Append Operation")
    plt.legend()
    plt.subplot(413)
    plt.plot(x,t_comprehension, label="List Comprehension Operation")
    plt.legend()
    plt.subplot(414)
    plt.plot(x,t_list, label="List Cretion Operation")
    plt.legend()
    plt.show()
```
