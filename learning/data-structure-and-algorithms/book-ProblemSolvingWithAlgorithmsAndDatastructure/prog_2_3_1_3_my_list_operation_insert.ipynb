List's insert  operation has a thetha(n) time complexity

We are using timeit module to find the complexity


```python
from timeit import Timer
import matplotlib.pyplot  as plt
```

```python
x = list(range(100,10000,100))
#y = [ list(range(1,i)) for i in x]
t_extend = []
```

```python

for i in x:
    l =  list(range(100001))
    #print le
    t1 = Timer(stmt="l.insert(i,5)",setup="from __main__ import l, i")
    t_extend.append(t1.timeit(number=1000))
    
```

```python
plt.figure()

plt.plot(x,t_extend,label="List insert operation")
plt.legend()
plt.show()

```

```python
plt.close()
```
